#  Researchers Data Dictionary - Uniform Data Set (UDS): uds3-rdd




```python
tables = camelot.read_pdf('uds3-rdd.pdf', pages='all')
tables
```

```python
for i in range(len(tables)):
    print( tables[i].df )
```

## "Revisions made to the UDS RDD" table

table 0 and table 1 are the same table

```python
first_table_columns = ["Date yyyy-mm-dd", "Description", "Form(s) affected", "Data element(s) affected"]
# {old_column: new_column for new_column, old_column in zip(first_table_columns, tables[0].df.columns)}
table1 = pd.concat([tables[0].df, tables[1].df], axis=0).iloc[1:]
table1.columns = first_table_columns
row_that_was_missing = pd.Series(["2015-12-03", "Variable name naccppam changed to naccppme", "D1", "naccppme/naccpam, naccppag"], index=table1.columns).to_frame().T
table1 = pd.concat([table1, row_that_was_missing], axis = 0)
table1 = table1.replace(r'\n',' ', regex=True) 
table1
```

## "Other data available at NACC" tables

tables on pages 7 (i=3) to 39 (i=34)

```python
for i in range(3,34):
    print( tables[i].df )
```

```python
second_tables_columns = list(tables[34].df.iloc[0])[0].split("\n")
table_list = []
table2 = pd.DataFrame()
for i in range(3,34):
    current_table = tables[i].df.iloc[1:]
    current_table.columns = second_tables_columns
    current_table = current_table.replace(r'\n',' ', regex=True) 
    table2 = pd.concat([table2, current_table], axis = 0)

table2
```

## Write both tables to excel

```python
output_file = './uds3-rdd.xlsx'
with pd.ExcelWriter(output_file, engine='xlsxwriter') as writer:
    # Write each DataFrame to a specific sheet
    table1.to_excel(writer, sheet_name='Other data available at NACC', index=False)
    table2.to_excel(writer, sheet_name='Revisions made to the UDS RDD', index=False)
```
